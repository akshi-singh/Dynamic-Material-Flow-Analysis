```python
import dynamic_stock_model as dsm # import the dynamic stock model libraryimport numpy as np
import numpy as np
import matplotlib.pyplot as plt
import openpyxl
```

```python
Cars_Datafile  = openpyxl.load_workbook(filename='C:\\Users\\Akshi\\OneDrive\\Documents\\MSC\\ODYM-master\\data\\MSC_DMFA_Data.xlsx')

Cars_Datasheet = Cars_Datafile['SC2_BS']

Years = [] #years 1950 to 2050
for m in range(2,103): # Read years
    Years.append(int(Cars_Datasheet.cell(m,1).value))

Population = [] #Global population, unit: 1, 1950 to 2050
for m in range(2,103): # Read years
    Population.append(int(Cars_Datasheet.cell(m,2).value))

CarOwnership = [] #car ownership rate per 1000 people, unit: 1, 1950 to 2050
for m in range(2,103): # Read years
    CarOwnership.append(Cars_Datasheet.cell(m,3).value)

CarLifetime = [] #average vehicle lifetime, unit: yr, 1950 to 2050
for m in range(2,103): # Read years
    CarLifetime.append(Cars_Datasheet.cell(m,4).value)    

CarAlpha = []  # shape
for m in range(2, 103):  # adjust range to your years
    CarAlpha.append(Cars_Datasheet.cell(m, 5).value)

CarBeta = []   # scale
for m in range(2, 103):  # adjust range to your years
    CarBeta.append(Cars_Datasheet.cell(m, 6).value)

CarAluContent = [] #Aluminium content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarAluContent.append(Cars_Datasheet.cell(m,7).value)    

CarSSContent = [] #Standatd steel Steel content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarSSContent.append(Cars_Datasheet.cell(m,8).value)

CarHSSContent = [] #HSS content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarHSSContent.append(Cars_Datasheet.cell(m,9).value)

CarFeContent = [] #Steel content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarFeContent.append(Cars_Datasheet.cell(m,10).value)

CarCuContent = [] #Steel content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarCuContent.append(Cars_Datasheet.cell(m,11).value)

CarZnContent = [] #Steel content of car, unit: kg/unit, 1950 to 2050
for m in range(2,103): # Read years
    CarZnContent.append(Cars_Datasheet.cell(m,12).value)



        
print(CarOwnership)
print(CarBeta)
```

```python
#With these data we can now define a dynamic stock model by creating an instance of the DynamicStockModel class and assigning time, inflow, and lifetime. 
#For the average product lifetime we chose a normally distributed lifetime with the mean of 30 years and a standard deviation of 30% of the mean.

TotalVehicleStock = [a*b/1000 for (a,b) in zip(Population,CarOwnership)] # unit: 1
print(TotalVehicleStock)
```

```python
#In stock-driven modelling, only the stock is known at first, 
#and we want to know the inflow, the outflow, and the cohort structure of both stock and outflow. 
#The DynamicStockModel class can do that.
#Below the dynamic stock model for the global passenger vehicle fleet is defined as instance of the DynamicStockModel class. 
#The list of years and the Total vehicle stock are passed on to that object, and the lifetime distribution is specified as well. 
#The parameters t and s are handed over as lists, the lt parameters as numpy arrays.
```

```python
GlobalVehicles_DSM = dsm.DynamicStockModel(t = Years, s = TotalVehicleStock, 
                                      lt = {'Type': 'Weibull', 
                                            'Shape': np.array(CarAlpha), 
                                            'Scale': np.array(CarBeta) })
CheckStr = GlobalVehicles_DSM.dimension_check()
print(CheckStr)
```

```python
#The central method for stock-driven modelling is compute_stock_driven_model(), 
#which contains a recursive calculation starting from the fist year to determine outflow, inflow, and stock by cohorts using the lifetime model and the mass balance. 
#The model is documented here: http://pubs.acs.org/doi/suppl/10.1021/es201904c/suppl_file/es201904c_si_001.pdf
```

```python
S_C, O_C, I = GlobalVehicles_DSM.compute_stock_driven_model()
# S_C: Stock by cohort
# O_C: Outflow by cohort
# I: inflow (new registration) of cars

O   = GlobalVehicles_DSM.compute_outflow_total() # Total outflow
DS  = GlobalVehicles_DSM.compute_stock_change()  # Stock change
Bal = GlobalVehicles_DSM.check_stock_balance()   # Vehicle balance

print(np.abs(Bal).sum()) # show sum absolute of all mass balance mismatches.
```

```python
SS_Stock_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarSSContent) # in kg, s_c has dimension year x cohort
Alu_Stock_ByCohort     = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarAluContent)   # in kg, s_c has dimension year x cohort
HSS_Stock_ByCohort     = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarHSSContent)   # in kg, s_c has dimension year x cohort
Fe_Stock_ByCohort     = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarFeContent)   # in kg, s_c has dimension year x cohort
Cu_Stock_ByCohort     = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarCuContent)   # in kg, s_c has dimension year x cohort
Zn_Stock_ByCohort     = np.einsum('tc,c->t',GlobalVehicles_DSM.s_c,CarZnContent)   # in kg, s_c has dimension year x cohort
# index t stands for year/time, index c for age-cohort

Steel_Outflow_ByCohort = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarSSContent) # in kg, o_c has dimension year x cohort
Alu_Outflow_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarAluContent)   # in kg, o_c has dimension year x cohort
HSS_Outflow_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarHSSContent)   # in kg, o_c has dimension year x cohort
Fe_Outflow_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarFeContent)   # in kg, o_c has dimension year x cohort
Cu_Outflow_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarCuContent)   # in kg, o_c has dimension year x cohort
Zn_Outflow_ByCohort   = np.einsum('tc,c->t',GlobalVehicles_DSM.o_c,CarZnContent)   # in kg, o_c has dimension year x cohort
# index t stands for year/time, index c for age-cohort
```
